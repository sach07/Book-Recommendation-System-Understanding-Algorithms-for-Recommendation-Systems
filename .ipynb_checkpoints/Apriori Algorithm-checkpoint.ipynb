```python
numItems=len(items)
numBaskets=len(baskets)
```

```python
numItems,numBaskets
```

```python
def support(itemset):
    basketsubset=baskets
    for items in itemset:
        basketsubset=[basket for basket in basketsubset if items in basket]
    return float(len(basketsubset))/float(numBaskets)
```

```python
support(['3','18'])
```

```python
SupportItems1=[]
minsupport=0.01
for item in range(numItems):
    itemset=str(item)
    if support(itemset)>minsupport:
        SupportItems1.append(str(item))
```

```python
SupportItems1
```

```python
import itertools

def apriori(i,SupportItems,assocrules,newSupportItems,minsupport,minconfidence):
    for itemset in itertools.combinations(SupportItems,i):
        itemset=list(itemset) 
        if support(itemset)>minsupport:
            for j in range(i):
                rule_to=itemset[j]
                rule_from=[x for x in itemset if x!=itemset[j]]
                confidence=support(itemset)/support(rule_from)
                if confidence>minconfidence:
                    assocrules.append((rule_from,rule_to))
                    #print(assocrules)
                    for x in itemset:
                        if x not in newSupportItems:
                            newSupportItems.append(x)
    return assocrules,newSupportItems
```

```python
minsupport=0.01
minconfidence=0.001
assocrules=[]
newSupportItems=[]
assocrules,SupportItems2=apriori(2,SupportItems1,assocrules,newSupportItems,minsupport,minconfidence)
```
